```python
def get_type_missing(df):
    # Creiamo un DataFrame vuoto dove salveremo alcune informazioni rilevanti
    df_types = pd.DataFrame()
    
    # Salviamo i tipi di dati di ciascuna colonna del DataFrame originale
    df_types['data_type'] = df.dtypes
    
    # Ciascuna colonna riceverà il conteggio dei valori mancanti (NaN)
    df_types['missing_values'] = df.isnull().sum()
    
    # Ordiniamo il DataFrame risultante in base ai valori mancanti in ordine decrescente
    return df_types.sort_values(by='missing_values', ascending=False)
```

### Riduzione delle classi maggioritarie (undersampling)

```python
from sklearn.utils import resample

# Otteniamo le frequenze di ogni classe
class_counts = df['Attack_type'].value_counts()
print("Distribuzione originale delle classi:")
print(class_counts)

# Troviamo la classe con minor numero di campioni
min_samples = class_counts.min()
balanced_dfs = []

# Per ogni classe, prendiamo un campione casuale di dimensione min_samples
for attack_type in df['Attack_type'].unique():
    # Prendiamo tutti i campioni di questa classe
    class_df = df[df['Attack_type'] == attack_type]
    
    # Se la classe ha più campioni del minimo, facciamo undersampling
    if len(class_df) > min_samples:
        class_df = resample(class_df,
                          replace=False,           # campionamento senza sostituzione
                          n_samples=min_samples,   # match numero di campioni
                          random_state=42)         # per riproducibilità
    
    balanced_dfs.append(class_df)

# Combiniamo tutti i DataFrame bilanciati
df_balanced_under = pd.concat(balanced_dfs)

# Verifichiamo la nuova distribuzione
print("\nNuova distribuzione delle classi dopo l'undersampling:")
print(df_balanced_under['Attack_type'].value_counts())

# Visualizziamo la nuova distribuzione con un grafico a torta
plt.figure(figsize=(10, 7))
attack_counts_balanced = df_balanced_under['Attack_type'].value_counts()
plt.pie(attack_counts_balanced, labels=attack_counts_balanced.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuzione dei tipi di attacco dopo undersampling')
plt.show()

print(f"\nDimensione originale del dataset: {len(df)}")
print(f"Dimensione del dataset dopo undersampling: {len(df_balanced_under)}")
```

### Sovracampionamento delle classi minoritarie (oversampling)

```python
from sklearn.utils import resample

# Otteniamo le frequenze di ogni classe
class_counts = df['Attack_type'].value_counts()
print("Distribuzione originale delle classi:")
print(class_counts)

# Troviamo la classe con maggior numero di campioni
max_samples = class_counts.max()
balanced_dfs = []

# Per ogni classe, facciamo oversampling fino a max_samples
for attack_type in df['Attack_type'].unique():
    # Prendiamo tutti i campioni di questa classe
    class_df = df[df['Attack_type'] == attack_type]
    
    # Se la classe ha meno campioni del massimo, facciamo oversampling
    if len(class_df) < max_samples:
        class_df = resample(class_df,
                          replace=True,            # campionamento con sostituzione
                          n_samples=max_samples,   # match numero di campioni
                          random_state=42)         # per riproducibilità
    
    balanced_dfs.append(class_df)

# Combiniamo tutti i DataFrame bilanciati
df_balanced_over = pd.concat(balanced_dfs)

# Verifichiamo la nuova distribuzione
print("\nNuova distribuzione delle classi dopo l'oversampling:")
print(df_balanced_over['Attack_type'].value_counts())

# Visualizziamo la nuova distribuzione con un grafico a torta
plt.figure(figsize=(10, 7))
attack_counts_balanced = df_balanced_over['Attack_type'].value_counts()
plt.pie(attack_counts_balanced, labels=attack_counts_balanced.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuzione dei tipi di attacco dopo oversampling')
plt.show()

print(f"\nDimensione originale del dataset: {len(df)}")
print(f"Dimensione del dataset dopo oversampling: {len(df_balanced_over)}")
```

### Vantaggi e Svantaggi
##### Undersampling
 - Pro:
    Minor costo computazionale
    Evita problemi di memoria
    Riduce il tempo di training
 - Contro:
    Perdita di informazioni potenzialmente utili
    Rischio di sottorappresentare la variabilità della classe maggioritaria
##### Oversampling
 - Pro:
    Mantiene tutte le informazioni originali
    Non perde esempi potenzialmente importanti
 - Contro:
    Rischio di overfitting sulle classi minoritarie
    Maggior costo computazionale
    Può creare copie identiche dei dati
    Scelta tra i Due
##### La scelta dipende da:
   Dimensione del dataset originale

   Sbilanciamento delle classi

   Risorse computazionali disponibili
   
   Importanza relativa di falsi positivi vs falsi negativi


## Training & Prediction

### Importiamo Bayes

```python
import numpy as np
from scipy.stats import norm

class GaussianNaiveBayes:
    """
    Implementazione del Naive Bayes Gaussiano per caratteristiche continue.
    
    Il Naive Bayes Gaussiano assume che le feature continue seguano una 
    distribuzione normale (gaussiana) all'interno di ogni classe.
    
    Parametri:
    ----------
    var_smoothing : float, default=1e-9
        Valore aggiunto alle varianze per evitare divisioni per zero.
    """
    
    def __init__(self, var_smoothing=1e-9):
        self.var_smoothing = var_smoothing
        self.class_priors = {}  # Probabilità a priori delle classi
        self.means = {}         # Media per ogni feature per ogni classe
        self.variances = {}     # Varianza per ogni feature per ogni classe
        self.classes = None     # Classi possibili
        self.n_features = None  # Numero di feature
        
    def fit(self, X, y):
        """
        Addestra il classificatore Naive Bayes Gaussiano.
        
        Parametri:
        ----------
        X : array-like, shape (n_samples, n_features)
            Matrice delle feature continue.
        y : array-like, shape (n_samples,)
            Vettore delle classi.
        
        Returns:
        --------
        self : object
        """
        X = np.asarray(X)
        y = np.asarray(y)
        
        n_samples, self.n_features = X.shape
        self.classes = np.unique(y)
        
        # Calcolo delle probabilità a priori delle classi
        for c in self.classes:
            self.class_priors[c] = np.mean(y == c)
            self.means[c] = {}
            self.variances[c] = {}
            
            # Calcolo media e varianza per ogni feature per questa classe
            X_c = X[y == c]
            
            for feature_idx in range(self.n_features):
                feature_values = X_c[:, feature_idx]
                
                # Calcolo media
                self.means[c][feature_idx] = np.mean(feature_values)
                
                # Calcolo varianza (con smoothing)
                self.variances[c][feature_idx] = np.var(feature_values) + self.var_smoothing
        
        return self
    
    def _calculate_gaussian_probability(self, x, mean, variance):
        """
        Calcola la densità di probabilità gaussiana.
        
        Parametri:
        ----------
        x : float
            Valore di cui calcolare la probabilità.
        mean : float
            Media della distribuzione gaussiana.
        variance : float
            Varianza della distribuzione gaussiana.
        
        Returns:
        --------
        float : Densità di probabilità calcolata.
        """
        # Implementazione diretta della formula della densità gaussiana
        exponent = np.exp(-((x - mean) ** 2) / (2 * variance))
        return exponent / np.sqrt(2 * np.pi * variance)
        
    def predict_log_proba(self, X):
        """
        Calcola il logaritmo delle probabilità per ciascuna classe.
        
        Parametri:
        ----------
        X : array-like, shape (n_samples, n_features)
            Matrice delle feature continue.
        
        Returns:
        --------
        array, shape (n_samples, n_classes) : Log-probabilità per ogni classe.
        """
        X = np.asarray(X)
        n_samples = X.shape[0]
        n_classes = len(self.classes)
        log_probs = np.zeros((n_samples, n_classes))
        
        for i in range(n_samples):
            for j, c in enumerate(self.classes):
                # Probabilità a priori (in log-spazio)
                log_prob = np.log(self.class_priors[c])
                
                # Somma dei logaritmi delle probabilità condizionate
                for feature_idx in range(self.n_features):
                    x_value = X[i, feature_idx]
                    mean = self.means[c][feature_idx]
                    variance = self.variances[c][feature_idx]
                    
                    # Calcoliamo la densità gaussiana e ne prendiamo il logaritmo
                    gaussian_prob = self._calculate_gaussian_probability(x_value, mean, variance)
                    
                    # Proteggiamoci da log(0)
                    if gaussian_prob > 0:
                        log_prob += np.log(gaussian_prob)
                    else:
                        log_prob += np.log(1e-10)  # Un valore molto piccolo
                
                log_probs[i, j] = log_prob
        
        return log_probs
    
    def predict_proba(self, X):
        """
        Restituisce le probabilità normalizzate delle classi.
        
        Parametri:
        ----------
        X : array-like, shape (n_samples, n_features)
            Matrice delle feature continue.
        
        Returns:
        --------
        array, shape (n_samples, n_classes) : Probabilità per ogni classe.
        """
        log_probs = self.predict_log_proba(X)
        
        # Normalizzazione numericamente stabile (log-sum-exp trick)
        log_prob_max = np.max(log_probs, axis=1).reshape(-1, 1)
        probs = np.exp(log_probs - log_prob_max)
        probs_sum = np.sum(probs, axis=1).reshape(-1, 1)
        
        return probs / probs_sum
    
    def predict(self, X):
        """
        Predice le classi per i campioni in X.
        
        Parametri:
        ----------
        X : array-like, shape (n_samples, n_features)
            Matrice delle feature continue.
        
        Returns:
        --------
        array, shape (n_samples,) : Le classi predette.
        """
        probs = self.predict_proba(X)
        return self.classes[np.argmax(probs, axis=1)]
    
    def score(self, X, y):
        """
        Calcola l'accuratezza del classificatore.
        
        Parametri:
        ----------
        X : array-like, shape (n_samples, n_features)
            Matrice delle feature continue di test.
        y : array-like, shape (n_samples,)
            Classi vere dei campioni di test.
        
        Returns:
        --------
        float : Accuratezza del classificatore.
        """
        return np.mean(self.predict(X) == y)
```

## Secondo Approccio - Domain Knowledge

Sfruttiamo le nostre conoscenze, studi, etc per selezionare le features


### Colonne e Spiegazioni

1. **`frame.time`**
    
    - _Motivo:_ Fornisce solo il timestamp della cattura del pacchetto. Il solo timing non indica un comportamento malevolo.
2. **`ip.src_host`**
    
    - _Motivo:_ I valori dell'host di origine possono essere facilmente falsificati (spoofing) o rimanere statici, offrendo scarso potere discriminante tra traffico benigno e malevolo.
3. **`ip.dst_host`**
    
    - _Motivo:_ Le informazioni sull'host di destinazione sono comuni sia al traffico normale che a quello di attacco, quindi non costituiscono un indicatore affidabile di un attacco.
4. **`arp.src.proto_ipv4`**
    
    - _Motivo:_ Questo dettaglio ARP fa parte della normale operatività del protocollo e non è correlato al comportamento di un attacco.
5. **`arp.dst.proto_ipv4`**
    
    - _Motivo:_ Simile al campo ARP di origine, questo è un attributo standard del protocollo che non segnala anomalie.
6. **`http.file_data`**
    
    - _Motivo:_ I dati dei file HTTP sono altamente variabili e possono essere cifrati, rendendoli un indicatore inconsistente di attacchi.
7. **`http.request.full_uri`**
    
    - _Motivo:_ Gli URI completi possono essere manipolati facilmente e non differenziano in modo affidabile le richieste benigne da quelle malevole.
8. **`icmp.transmit_timestamp`**
    
    - _Motivo:_ I timestamp di trasmissione in ICMP si concentrano sul timing di rete, non su comportamenti indicativi di un attacco.
9. **`http.request.uri.query`**
    
    - _Motivo:_ Le stringhe di query sono altamente variabili e possono essere alterate senza necessariamente indicare un intento malevolo.
10. **`tcp.options`**
    
    - _Motivo:_ Si tratta di dettagli di configurazione TCP di basso livello che generalmente aggiungono rumore piuttosto che segnali utili per il rilevamento.
11. **`tcp.payload`**
    
    - _Motivo:_ I payload possono essere cifrati o inconsistenti, riducendo il loro valore come feature robusta per identificare gli attacchi.
12. **`tcp.srcport`**
    
    - _Motivo:_ I numeri di porta di origine sono spesso assegnati dinamicamente e non fungono da marcatori affidabili per gli attacchi.
13. **`tcp.dstport`**
    
    - _Motivo:_ Le porte di destinazione tendono ad essere standard per molti servizi e non forniscono un valore discriminante sufficiente.
14. **`udp.port`**
    
    - _Motivo:_ Come le porte TCP, i numeri di porta UDP sono tipicamente standard e non abbastanza specifici da indicare attacchi.
15. **`mqtt.msg`**
    
    - _Motivo:_ Il contenuto dei messaggi MQTT può essere altamente variabile, a volte cifrato o troppo breve, rendendolo inaffidabile per il rilevamento.
16. **`icmp.unused`**
    
    - _Motivo:_ Questo campo non è utilizzato dal protocollo e non trasporta informazioni significative per il rilevamento di attacchi.
17. **`http.tls_port`**
    
    - _Motivo:_ La porta TLS è un dettaglio statico del protocollo e non fornisce informazioni sull'attività malevola.
18. **`dns.qry.type`**
    
    - _Motivo:_ I tipi di query DNS sono comuni nel traffico legittimo e non differenziano il comportamento normale da quello correlato agli attacchi.
19. **`dns.retransmit_request_in`**
    
    - _Motivo:_ Gli indicatori di ritrasmissione riflettono generalmente problemi di rete tipici piuttosto che essere specifici dei pattern di attacco.
20. **`mqtt.msg_decoded_as`**
    
    - _Motivo:_ La rappresentazione decodificata non aggiunge valore discriminante extra oltre al messaggio MQTT grezzo.
21. **`mbtcp.trans_id`**
    
    - _Motivo:_ In Modbus/TCP, l'ID di transazione è un identificatore arbitrario utilizzato per abbinare richieste e risposte, senza alcuna relazione con il comportamento di un attacco.
22. **`mbtcp.unit_id`**
    
    - _Motivo:_ Simile all'ID di transazione, l'ID unità è un identificatore costante che non contribuisce al rilevamento degli attacchi.
23. **`http.request.method`**
    
    - _Motivo:_ I metodi HTTP (ad es. GET, POST) sono standard in molte comunicazioni e raramente indicano un intento malevolo.
24. **`http.referer`**
    
    - _Motivo:_ L'header referer può essere facilmente falsificato e non è correlato in modo affidabile ai pattern di attacco.
25. **`http.request.version`**
    
    - _Motivo:_ I numeri di versione HTTP sono spesso costanti per una data comunicazione e aggiungono scarso potere discriminante.
26. **`dns.qry.name.len`**

29. **`mqtt.topic`**
    
    - _Motivo:_ I topic possono essere scelti arbitrariamente dai publisher e quindi non segnalano in modo affidabile attività malevole.

33. **`tcp.len`**

34. **`udp.stream`**

35. **`udp.time_delta`**

36. **`dns.qry.name`**

37. **`dns.qry.qu`**

#### Riapplichiamo l'undersampling

```python
from sklearn.utils import resample

# Otteniamo le frequenze di ogni classe
class_counts = df['Attack_type'].value_counts()
print("Distribuzione originale delle classi:")
print(class_counts)

# Troviamo la classe con minor numero di campioni
min_samples = class_counts.min()
balanced_dfs = []

# Per ogni classe, prendiamo un campione casuale di dimensione min_samples
for attack_type in df['Attack_type'].unique():
    # Prendiamo tutti i campioni di questa classe
    class_df = df[df['Attack_type'] == attack_type]
    
    # Se la classe ha più campioni del minimo, facciamo undersampling
    if len(class_df) > min_samples:
        class_df = resample(class_df,
                          replace=False,           # campionamento senza sostituzione
                          n_samples=min_samples,   # match numero di campioni
                          random_state=42)         # per riproducibilità
    
    balanced_dfs.append(class_df)

# Combiniamo tutti i DataFrame bilanciati
df_balanced_under = pd.concat(balanced_dfs)

# Verifichiamo la nuova distribuzione
print("\nNuova distribuzione delle classi dopo l'undersampling:")
print(df_balanced_under['Attack_type'].value_counts())

# Visualizziamo la nuova distribuzione con un grafico a torta
plt.figure(figsize=(10, 7))
attack_counts_balanced = df_balanced_under['Attack_type'].value_counts()
plt.pie(attack_counts_balanced, labels=attack_counts_balanced.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuzione dei tipi di attacco dopo undersampling')
plt.show()

print(f"\nDimensione originale del dataset: {len(df)}")
print(f"Dimensione del dataset dopo undersampling: {len(df_balanced_under)}")
```
